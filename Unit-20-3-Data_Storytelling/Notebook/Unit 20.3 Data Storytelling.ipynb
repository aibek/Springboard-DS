## 3. Explore Data

### So, what are even looking at? Show us these PAD districts and refineries in the U.S

## 4. Hypothesis and Insights

The first thing we notice is that the PAD districts with the highest concentration of refineries, i.e. TX (PAD-3), LA (PAD-3), and CA (PAD-5) are NOT in the list of "high crack spread" candidates.

The 'highest gross margins' are shown in PAD-4 and PAD-2.

The obviouis reason for that is probably a higher supply to demand ratio, compared to other districts. Since refineries target as many destination markets, as it's economically viable, let us look at what really drives the "supply to demand" pressure, i.e. distances from refineries to target markets.

#### Distribution of cumulative "Refinery to Destination Market" distances - by PAD districts

```python
distances = data.iloc[:,np.r_[41:48]]
distances.hist(figsize=(15,7))
plt.subplots_adjust(hspace=0.5);
```

The histograms confirm only a part of the above hypothesis - refineries in the PAD-3 do primarily have close proximity (500-750 miles) to target markets and hence, create the highest supply pressure. 

However, gross margins of refineries in the PAD-5, particualrly, in CA, seem to struggle for different reasons - there isn't much supply pressure and it's probably due to considerably lower production capacity per population, compared to those of TX and LA refineries.

Refineries in the PAD districts 4 and 2 have a low concentration, relative to closest markets and hence, low supply pressure - we can guess that is the reason for potentially higher crack spread for the local production.

Hence, the higher profit margins? It looks that way - let's consider the other key feature, the Product Yields.

#### Product Yield impact:

Product Yields = breakdown of various petroleum products, which may be extracted from processing one unit (say, 1 U.S. barrel) of crude oil. The examples of refined petroleum products include Gasoline, Diesel, Fuel Oil, residue - each having a differing market value.

The ratio of refined products depends on the type of crude oil, which varies in quality and characteristic, as well as processing capacity of a given refinery - e.g. 2 refineries make intake the same type of crude oil, yet output products, which would vary in quality and hence, market value.

The light products, such as Gasoline and Diesel are valued a lot higher than Residue, so we may use the Residue Yield as a reference metric on how well of a job a given refinery is doing in extracting high-value products as Gasoline and Diesel.

##### Product Yield Correlations to Crack Spread values:

Let's look at a scatter plots of Residue Yields to crack spread values - by PAD District:

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x='RFY-PAD-District', y='SPECF_RFY_CRCK', hue='RESID_Yield_PAD', data=rfy_df)
```

OK, another hypothesis is confirmed! We can see a clear correlation between the Crack Spreads and Residue Yield - the lower it is, the higher portion of products consists of Gasoline and Diesel, ultimately leading to higher revenues and - crack spreads.

Look at the PAD district distributions - refineries in the PAD-4, PAD-2-3, PAD-2-1, and PAD-2-2 are the leaders, which is consistent with our initial observation.

#### Does the type of crude have to do much with the yeilds and crack spreads?

The feature analysis ranked Crude API at a lower end of the feature importance. However, we would like to see what its correlation is with the Product Yeilds.

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x='RESID_Yield_PAD', y='Crude_API', hue='SPECF_RFY_CRCK', data=rfy_df)
```

Well, yes - although refineries with advanced configurations may extract higher volumes of Gasoline and Diesel products than the simpler ones, overall, the higher API means more light products than the lower API. The scatterplot above demonstrates that - using Residude Yield as a reference metric.

#### Now, just for a good measure, let's look at the Crude API impact on the crack spreads by PAD district:

```python
sns.relplot(x='RESID_Yield_PAD', y='SPECF_RFY_CRCK', data=rfy_df, kind='scatter',
            hue='Crude_API', col='RFY-PADD', col_wrap=2)
```

OK, these charts show that how refineries in the PAD-4 and PAD-2 districts have an upper hand in making better profit margins than those in PAD-1 and PAD-3 - purely based on the availability of Crude Oil with API of 35-40, because of their proximity to the source and the perceived WTI discounts they would be getting.

## Conclusion

Our analysis has confirmed that selecting PAD Districts PAD-4, PAD-2-1, PAD-2-2, or PAD-2-3 would be most optimal, based on the following correlations, visualized above:
1. The higher Crude Oil Price Differential (Discount), the lower is the cost of acquired crude oil
2. The proximity to largest markets is important - for as long as the ratio between the Supply Pressure and Market Resistance is closer to the values, found in PADs 4 and 2
