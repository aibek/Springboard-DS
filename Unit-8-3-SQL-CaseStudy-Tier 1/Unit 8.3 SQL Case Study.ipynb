```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT *
        FROM FACILITIES
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

### Q10: Produce a list of facilities with a total revenue less than 1000.
         The output of facility name and total revenue sorted by revenue.
         Remember that there is a different cost for guests and members!

```python
def Q10_task (conn):
    """
    Query to answer Question 10 of the Case Study
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        WITH f_rev_list AS (
        SELECT f.name,
            CASE WHEN b.memid=0 THEN b.slots*f.guestcost
                 WHEN b.memid!=0 THEN b.slots*f.membercost
                 END AS f_revenue
        FROM BOOKINGS AS b
        LEFT JOIN FACILITIES AS f
        ON b.facid = f.facid)
        
        SELECT f_rev_list.name, SUM(f_revenue)
        FROM f_rev_list
        GROUP BY f_rev_list.name
        HAVING SUM(f_revenue) < 1000
        ORDER BY f_revenue;
        """
    cur.execute(query1)
    rows = cur.fetchall()
    for row in rows:
        print(row)

def Q10_execute ():
    database = "sqlite_db_pythonsqlite.db"
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("Question 10 Query")
        print("b.id|b.facid|f.Name|f_revenue ")
        Q10_task(conn)
        
Q10_execute()
```

### Q11: Produce a report of members and who recommended them in alphabetic surname, firstname order

```python
'The outcome of this query excludes members, who were not referred by anyone'

def Q11_task (conn):
    """
    Query to answer Question 11 of the Case Study
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT
            (m1.surname || ' ' || m1.firstname) AS Member_Name,
            (m2.surname || ' ' || m2.firstname) AS Recommended_By_Name
        FROM
            MEMBERS AS m1, MEMBERS AS m2
        WHERE
            m1.recommendedby = m2.memid
        ORDER BY Member_Name, Recommended_By_Name
        """
    cur.execute(query1)
    rows = cur.fetchall()
    for row in rows:
        print(row)

def Q11_execute ():
    database = "sqlite_db_pythonsqlite.db"
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("Question 11 Query")
        Q11_task(conn)
        
Q11_execute()
```

### Q12: Find the facilities with their usage by member, but not guests

```python
def Q12_task (conn):
    """
    Query to answer Question 12 of the Case Study
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT
            b.facid, f.name, b.memid,
            (m.surname || ' ' || m.firstname) AS Member_Name,
            COUNT(CASE WHEN b.memid!=0 THEN 0 END) AS Member_Usage
        FROM Bookings AS b
        LEFT JOIN Facilities AS f
        ON b.facid = f.facid
        LEFT JOIN Members AS m
        ON b.memid = m.memid
        WHERE b.memid !=0
        GROUP BY b.facid, b.memid
        ORDER BY Member_Name
        """
    cur.execute(query1)
    rows = cur.fetchall()
    for row in rows:
        print(row)

def Q12_execute ():
    database = "sqlite_db_pythonsqlite.db"
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("Question 12 Query")
        Q12_task(conn)
        
Q12_execute()
```

### Q13: Find the facilities usage by month, but not guests

```python
def Q13_task (conn):
    """
    Query to answer Question 13 of the Case Study
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT b.facid, f.name,
            strftime('%m', b.starttime) AS Usage_Month,
            COUNT(CASE WHEN b.memid!=0 THEN 0 END) AS Member_Usage
        FROM BOOKINGS AS b
        LEFT JOIN FACILITIES AS f
        ON b.facid = f.facid
        LEFT JOIN Members AS m
        ON b.memid = m.memid
        WHERE b.memid !=0
        GROUP BY b.facid, Usage_Month
        ORDER BY b.facid, Usage_Month
        """
    cur.execute(query1)
    rows = cur.fetchall()
    for row in rows:
        print(row)

def Q13_execute ():
    database = "sqlite_db_pythonsqlite.db"
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("Question 13 Query")
        Q13_task(conn)
        
Q13_execute()
```
